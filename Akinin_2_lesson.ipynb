# Численные последолвательности 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Зададим последовательность как
$$ a_n = \frac{1}{n^2+4n+31} $$

Тогда можно задать последовательную сумму
$$ S_N = \Sigma_{n=1}^N(a_n) $$

Задача: 

1) Найти предел последовательности сумм

2) Вывести график $a_n$

3) Вывести график $S_N$

### Создаем функцию рассчёта последовательности

```python
def func(n):
    return 1/(n**2 + 4*n + 31)
```

### Рассчитаем последовательность $a_n$ до того элемента, котогрый меньше чем $\varepsilon$

```python
eps = 1e-4
```

```python
start_n = 1

n_span = [start_n,]
a_n = [func(start_n),]

n = start_n
while a_n[-1] > eps:
    n+=1
    a_n.append(func(n))
    n_span.append(n)
```

```python
plt.plot(n_span, a_n, color='r', label='последовательность', marker='.')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('n')
plt.ylabel('a(n)')
plt.grid(True)
plt.legend()
```

```python
s_n = [np.sum(a_n[0:n]) for n in range(len(a_n))]
print(len(a_n))
```

```python
plt.plot(n_span, s_n, color='g', label='последовательная сумма', marker='.')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('n')
plt.ylabel('s(n)')
plt.grid(True)
plt.legend()
```

### Выведем предел последовательных сумм $S_n$

```python
print(f'Последовательная сумма сходится к {s_n[-1]}')
```
