# Численное дифференциирование методом конечных приращений

$$ y(x+h) = y(x) + y'(x)\cdot h + \frac{y''(x)}{2}\cdot h^2 + \frac{y'''(x)}{6}\cdot h^3 + ... $$

$$ y'(x) = \frac{y(x+h)-y(x)}{h} + O(h) $$
$$ y'(x) = \frac{y(x)-y(x-h)}{h} + O(h) $$
$$y'(x) = \frac{y(x+h)-y(x-h)}{2h} + O(h^2) $$

Задание: есть функция арктангенс, надо 1) построить саму функ, 2) производную аналитическую, 3) построить численную производную с оптисальным шагом, 4) продемонстрировать ошибку численного метода

```python
import numpy as np
import matplotlib.pyplot as plt
```

## График функции (тангенс)

```python
x_start, x_end, x_num = 0, 10, 100
x_span = np.linspace(x_start, x_end, x_num)
y = np.arctan(x_span)

plt.plot(x_span, y, label='arctan', color='r')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.legend()
plt.grid(True)
```

## Аналитическая производная функции

```python
def arctan_deriv(x):
    return 1/(x**2 + 1)
```

```python
dy_dx_analitic = arctan_deriv(x_span)

plt.plot(x_span, dy_dx_analitic, label='analitic arctan derivative', color='g')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.legend()
plt.grid(True)
```

## Численная производная функции
Вычислим оптимальный шаг $h_{\text{opt}}$ при численном вычислении производной

```python
M_0 = np.max(y)
M_2 = 1
eps_mash = 1e-16

h_opt = 2*np.sqrt(M_0*eps_mash/M_2)
```

```python
dy_dx_numeric = np.zeros(x_num)

for i, x in enumerate(x_span):
    dy_dx_numeric[i] = (np.arctan(x+h_opt) - np.arctan(x))/h_opt

plt.plot(x_span, dy_dx_numeric, label='numeric arctan derivative', color='b')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.legend()
plt.grid(True)
```

## Разность между численной и аналитический производной (ошибка)

```python
dy_dx_error = dy_dx_analitic - dy_dx_numeric

plt.plot(x_span, dy_dx_error, label='error between analitic and numeric', color='orange')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.legend()
plt.grid(True)
```
